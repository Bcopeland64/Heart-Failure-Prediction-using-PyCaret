```python
from pycaret.classification import *
clf1 = setup(df, target='HeartDisease')
```

```python
compare_models()
```

```python
et = create_model('et')
```

```python
tuned_et = tune_model(et, optimize = "AUC")
```

```python
evaluate_model(tuned_et)
```

```python
final_et = finalize_model(tuned_et)
```

```python
print(final_et)
```

```python
predictions = predict_model(final_et, data = df)
```
